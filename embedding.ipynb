```python

import openai
import os
import yaml
import json
```

```python
with open("config.yaml", "r") as stream:
    try:
        PARAM = yaml.safe_load(stream)
    except yaml.YAMLError as exc:
        print(exc)
```

```python
openai.api_key  = PARAM['openai_api']
client = openai.OpenAI(api_key = PARAM['openai_api'])

def get_embedding(text, model="text-embedding-3-small"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding

```

```python
content = ""
undefined_processed = False
undefined_embedding = []
for line in open("drugs/disease_4.json", "r"):
    value = json.loads(line)
    embedding = []
    definition = ""
    print (value["cui"])

    if "definition" in value and value["definition"] is not None and len(value["definition"]) > 0:
        definition = value["definition"]
        embedding = get_embedding(value["definition"])
    else:
        if not undefined_processed:
            undefined_processed = True
            definition = "undefined"
            undefined_embedding = get_embedding(definition)
        
        definition = "undefined"
        embedding = undefined_embedding
    content += json.dumps({"cui": value["cui"], "name": value["name"], "definition": definition, "definitionEmbedding": embedding}) + "\n"
```

```python
with open("drugs/disease_4_embedding.json", "w") as f:
    f.write(content)
```
